## `if` statement

```
if condition_1:
    do_stuff_1
elif condition_2:
    do_stuff_2
# ...
else:
    do_stuff_n
```

**Esempio**

````python
x = 5

if x > 5:
    print("x è maggiore di 5")
elif x == 5:
    print("x è proprio 5")
else:
    print("x è minore di 5")
````

## `for` loop

```
for variable in iterable:
    do_stuff
```

**Esempio**

````python
for x in (1,2,3):
    print(3 * x)
````

## `while` loop

```
while condition:
    do_stuff
```


**Esempio**

````python
x = 1

print(x)

while x < 10:
    x = 2 * x
    if x == 8:
        print("x è 8, mi fermo")
        break
    
print(x)
````

### `try`, `except`, `pass`

```
try:
    do_stuff
except this_happen:
    do_something_else
```

````python
x = 1
y = 0

try:
    print(x/y)
except ZeroDivisionError:
    print("non posso dividere per 0...")
````
